Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Pre Processing

Loading the Data

```python
#loading the data from csv file
raw_mail_data = pd.read_csv('/Users/gautamparmar/Desktop/AIRepositories/ML Project/Spam Mail Prediction/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
#replacing the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
#prinitng first 5 rows
raw_mail_data.head()
```

```python
#checking the number of rows and coloumns in the data frame
mail_data.shape
```

Label encoding, converting ham and spam to the values of 1 and 0

```python
#label spam mail as - 0 and Ham mail as 1

mail_data.loc[mail_data['Category'] == 'spam', 'Category' ,] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category' ,] = 1
```

```python
#separating the data as text and labels

X = mail_data['Message']

Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Training - Testing - Split

```python
#splitting the data into training data and test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

Feature Extraction

```python
#transforming the text data i.e messages in feature vectors(numerical values) to use as imput data

feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)


X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)


#COnvert Y train and Y test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Training the Model

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with the Training Data
model.fit(X_train_features, Y_train)
```

Evaluating the Trained Model

```python
#prediction on training data
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on Training Data is =', accuracy_on_training_data)
```

```python
#Prediction on Test Data
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on Test Data = ', accuracy_on_test_data)
```

Building a Perdective System

```python
input_mail = ["I HAVE A DATE ON SUNDAY WITH WILL!!"]

#convert text to feature vectors

input_data_features = feature_extraction.transform(input_mail)

#making Predictions

predictiom = model.predict(input_data_features)
print(predictiom)

if predictiom[0] == 1:
    print('It is an Ham Mail')
else:
    print('Spam Mail')
```
